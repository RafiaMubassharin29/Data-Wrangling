```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv(r"C:\Users\HP\Downloads\FloodData1.csv")
df1.head()
```

```python
df1 = df1.drop("Sl", axis = 1)
```

```python
df2 = pd.read_csv(r"C:\Users\HP\Downloads\FloodData2.csv")
df2.head()
```

```python
# separate station_name and station_code
df2['Station_name'] = df2['Station'].str.extract(r'([a-zA-Z]+)')
df2['Station_code'] = df2['Station'].str.extract(r'(\d+)')
```

```python
# drop serial no
df2 = df2.drop('Station', axis = 1)
```

```python
df2 = df2[['Station_name','Year','Month','Maximum Temperature (Deg.Cel)',
     'Minimum Temperature (Deg.Cel)','Total Rainfall (mm)','Humidity (percent)',
    'Wind Speed (m/s)','Cloud Coverage (Octs)','Sunshine (Hours)','Station_code']]
```

```python
df2.sample(5)
```

```python
# match the column names with df1
df2 = df2.rename(columns = {'Station_name':'Station_Names',
                      'Year':'Year', 'Month':'Month', 'Maximum Temperature (Deg.Cel)': 'Max_Temp',
                      'Minimum Temperature (Deg.Cel)': 'Min_Temp', 'Total Rainfall (mm)' : 'Rainfall',
                      'Humidity (percent)': 'Relative_Humidity', 'Wind Speed (m/s)': 'Wind_Speed', 
                            'Cloud Coverage (Octs)': 'Cloud_Coverage',
                      'Sunshine (Hours)': 'Bright_Sunshine', 'Station_code' : 'Station_Number'})
```

```python
df1['Station_Names'].unique()
```

```python
df2['Station_Names'].unique()
```

```python
# drop the extra 2 station
df2 = df2[~df2['Station_Names'].isin(['Chuadanga', 'Syedpur', 'Srimongal'])]
```

```python
df2['Station_Names'].unique()
```

```python
# match the station names with df1
def replace_station_name(station):
    if station == 'Ambagan':
        return 'Chittagong (City-Ambagan)'
    elif station == 'Chittagong':
        return 'Chittagong (IAP-Patenga)'
    elif station == 'M':
        return 'Maijdee Court'
    elif station == 'Cox':
        return "Cox's Bazar"
    else:
        return station
```

```python
df2['Station_Names'] = df2['Station_Names'].apply(replace_station_name)
```

```python
df2['Station_Names'].unique()
```

```python
# sort values according to station_names, year and month
df2 = df2.sort_values(['Station_Names', 'Year', 'Month'])
```

```python
# resetting the index
df2 = df2.reset_index()
```

```python
df2.head()
```

```python
df2 = df2.drop('index', axis = 1)
```

```python
df2.head()
```

```python
# droped the station_number because the number is not matching with df1
df2 = df2.drop('Station_Number', axis = 1)
```

```python
# save df2 in a new csv file
df2.to_csv('clean_df2.csv', encoding='utf-8', header=True, index=False)
```

```python
# merge df1 and df2
merge_file = pd.concat([df1, df2], ignore_index = True)
```

```python
# Create a mapping of Station Name to Station Code from df1
station_code_map = df1.set_index('Station_Names')['Station_Number'].to_dict()
```

```python
# Fill the NaN values in 'Station Code' in merged_df based on the mapping
merge_file['Station_Number'] = merge_file['Station_Number'].fillna(merge_file['Station_Names'].map(station_code_map))
```
